## 課題1
**課題1-1:** 最小二乗法の説明において，$N$ 個の離散データ $\{x_\alpha \}^N_{\alpha=1}$ が与えられ，関数 $f(x), g(x)$ に対して，内積とノルムは次のように定義されていた．

内積：

$$
\left (f,g \right ) = \sum_{\alpha=1}^N f \left(x_\alpha \right) g \left(x_\alpha \right)
$$

ノルム：

$$
\left \| f \right \| = \sqrt{\left (f,f\right )} = \sqrt{\sum^{N}_{\alpha=1} f \left (x_\alpha\right )^2}
$$

この演算が内積の公理を満たすことを確かめよ（手計算もしくはSympy)．

内積の正値性について
$$
\begin{aligned}
(f, f) & =\sum_{\alpha=1}^N f\left(x_\alpha\right) f\left(x_\alpha\right) \\
& =\sum_{\alpha=1}^N f\left(x_\alpha\right)^2 \geq 0
\end{aligned}
$$
より，内積の正値性を満たす．

内積の対称性について
$$
\begin{aligned}
(f, g) & =\sum_{\alpha=1}^N f\left(x_\alpha\right) g\left(x_\alpha\right) \\
& =\sum_{\alpha=1}^N g\left(x_\alpha\right) f\left(x_\alpha\right) \\
& =(g, f)
\end{aligned}
$$
より，内積の対称性を満たす．

内積の線形性について

$$
\begin{aligned}
\left(c_1 f_1+c_2 f_2, g\right) & =\sum_{\alpha=1}^{N v}\left(c_1 f_1\left(x_\alpha\right)+c_2 f_2\left(x_\alpha\right)\right) g\left(x_\alpha\right) \\
& =\sum_{\alpha=1}^N c_1 f_1\left(x_\alpha\right) g\left(x_\alpha\right)+\sum_{\alpha=1}^N c_2 f_2\left(x_\alpha\right) g\left(x_\alpha\right) \\
& =c_1 \sum_{\alpha=1}^N f_1\left(x_\alpha\right) g\left(x_\alpha\right)+c_2 \sum_{\alpha=1}^N f_2\left(x_\alpha\right) g\left(x_\alpha\right) \\
& =c_1\left(f_1, g\right)+c_2\left(f_2, g\right)
\end{aligned}
$$

より，内積の線形性を満たす．

以上より，この演算は内積の公理を満たす．

**課題1-2:** $n$ 次元列ベクトル $\mathbf{a}, \mathbf{b}$ の内積とノルムは次のように定義されていた．

内積：

$$
\left (\mathbf{a}, \mathbf{b} \right ) = \sum_{i=1}^n a_i b_i
$$

ノルム：

$$
\left \| \mathbf{a} \right \| = \sqrt{\left ( \mathbf{a}, \mathbf{a} \right )} = \sqrt{\sum^{n}_{i=1} a_i^2}
$$

この演算が内積の公理を満たすことを確かめよ（手計算もしくはSympy)．

内積の正値性について
$$
\begin{aligned}
(\mathbf{a}, \mathbf{a}) & =\sum_{i=1}^n a_i a_i \\
& =\sum_{i=1}^n a_i^2 \geq 0
\end{aligned}
$$
より，内積の正値性を満たす．

内積の対称性について
$$
\begin{aligned}
(\mathbf{a}, \mathbf{b}) & =\sum_{i=1}^n a_i b_i \\
& =\sum_{i=1}^n b_i a_i \\
& =(\mathbf{b}, \mathbf{a})
\end{aligned}
$$
より，内積の対称性を満たす．

内積の線形性について
$$
\begin{aligned}
\left(c_1 \mathbf{a}_1+c_2 \mathbf{a}_2, \mathbf{b}\right) & =\sum_{i=1}^n\left(c_1 a_{1 i}+c_2 a_{2 i}\right) b_i \\
& =\sum_{i=1}^n c_1 a_{1 i} b_i+\sum_{i=1}^n c_2 a_{2 i} b_i \\
& =c_1 \sum_{i=1}^n a_{1 i} b_i+c_2 \sum_{i=1}^n a_{2 i} b_i \\
& =c_1\left(\mathbf{a}_1, \mathbf{b}\right)+c_2\left(\mathbf{a}_2, \mathbf{b}\right)
\end{aligned}
$$
より，内積の線形性を満たす．

以上より，この演算は内積の公理を満たす．

**課題1-３:** 区間 $[a,b]$ 上の連続関数 $f(x), g(x)$ に対して内積とノルムは次のように定義されていた．

内積：

$$
\left ( f, g \right ) = \int_{a}^{b} f\left(x\right) g\left(x\right) dx
$$

ノルム：

$$
\left \| \mathbf{f} \right \| = \sqrt{\left ( f, f  \right )} = \sqrt{ \int_{a}^{b} f\left(x\right) ^2 dx}
$$

この演算が内積の公理を満たすことを確かめよ（手計算もしくはSympy)．

内積の正値性について

$$
\left ( f, f \right ) = \int_{a}^{b} f\left(x\right)^2 dx \geq 0
$$

より，内積の正値性を満たす．

内積の対称性について，

$$
\left ( f, g \right ) = \int_{a}^{b} f\left(x\right) g\left(x\right) dx 
= \int_{a}^{b} g\left(x\right) f\left(x\right) dx 
= \left ( g, f \right )
$$

より，内積の正値性を満たす．

内積の線形性について，

$$
\begin{aligned}
\left ( \lambda_1 f + \lambda_2 g, h \right ) &= \int_{a}^{b} \left \{ \lambda_1 f\left(x\right) + \lambda_2 g\left(x\right) \right \} h\left(x\right)dx \\
&= \int_{a}^{b} \left \{ \lambda_1 f\left(x\right) h\left(x\right) + \lambda_2 g\left(x\right) h\left(x\right)\right \} dx \\
&=  \lambda_1  \int_{a}^{b} f\left(x\right) h\left(x\right)dx + \lambda_2  \int_{a}^{b} g\left(x\right) h\left(x\right) dx \\
&= \lambda_1 \left ( f, h \right ) + \lambda_2 \left ( g, h \right )
\end{aligned}
$$

より，内積の線形性を満たす．

以上より，この演算は内積の公理を満たす．
