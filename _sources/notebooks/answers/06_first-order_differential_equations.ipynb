## 課題1
**課題1-1:** 次の微分方程式の一般解を求めよ．（手計算）

$$
\frac{dy}{dx} + \frac{y}{x} = xy^2
$$

これは $n=2$ のベルヌーイ型であるので， $z=y^{-1}$ として変数変換する.
$$
\begin{aligned}
\frac{d z}{d x} & =\frac{d}{d x} y^{-1} \\
& =-y^{-2} \frac{d y}{d x}
\end{aligned}
$$

これを使って元の微分方程式を変数変換の関数 $z$ と $\frac{d z}{d x}$ で書き換える. そのために両辺を $y^2$ で割り，変数変換の結果を代入すると
$$
\begin{aligned}
\frac{d y}{d x}+\frac{y}{x} & =x y^2 \\
\frac{1}{y^2} \frac{d y}{d x}+\frac{1}{y x} & =x \\
-\frac{d z}{d x}+\frac{z}{x} & =x \\
\frac{d z}{d x}-\frac{z}{x} & =-x
\end{aligned}
$$

となる. これは $p(x)=-\frac{1}{x}, q(x)=-x$ の非同次微分方程式であるので， $z=\frac{1}{y}$ に注意して一般解を導出する.
まず積分因子 $e^{\int p(x) d x}$ を計算する. ここで $p(x)=-\frac{1}{x}$ より，
$$
e^{\int p(x) d x}=e^{\int-\frac{1}{x} d x}=\frac{1}{x}
$$

となる. 両辺に積分因子をかけると,
$$
\begin{aligned}
\left(\frac{d z}{d x}-\frac{z}{x}\right) \frac{1}{x} & =-1 \\
\frac{d z}{x d x}-\frac{z}{x^2} & =-1
\end{aligned}
$$

左辺に着目すると関数 $\frac{z}{x}$ の微分になっているので両辺を積分し,左辺の積分定数 $C_1$ と右辺の積分定数 $C_2$ が存在するがまとめて $C=C_1+C_2$ とし, $z$ について整理したのち, $y$ について整理する.
$$
\begin{aligned}
\frac{d}{d x}\left(\frac{z}{x}\right) & =-1 \\
\frac{z}{x} & =-x+C \\
z & =-x^2+C x \\
\frac{1}{y} & =-x^2+C x \\
y & =-\frac{1}{x^2+C x}
\end{aligned}
$$

となる.

**課題1-2:** 次の微分方程式の一般解を求めよ．（手計算）

$$
x\frac{dy}{dx} -y = y^3
$$

これは $n=3$ のベルヌーイ型であるので, $z=y^{-2}$ として変数変換する.
$$
\begin{aligned}
\frac{d z}{d x} & =\frac{d}{d x} y^{-2} \\
& =-2 y^{-3} \frac{d y}{d x}
\end{aligned}
$$

これを使って元の微分方程式を変数変換の関数 $z$ と $\frac{d z}{d x}$ で書き換える. そのために両辺を $y^3$ で割り，変数変換の結果を代入すると

$$
\begin{aligned}
x \frac{d y}{d x}-y & =y^3 \\
\frac{x}{y^3} \frac{d y}{d x}-\frac{1}{y^2} & =1 \\
-\frac{x}{2} \frac{d z}{d x}-z & =1 \\
\frac{d z}{d x}+\frac{2 z}{x} & =-\frac{2}{x}
\end{aligned}
$$

となる. これは $p(x)=\frac{2}{x}, q(x)=-\frac{2}{x}$ の非同次微分方程式であるので， $z=\frac{1}{y^2}$ に注意して一般解を導出する. まず積分因子 $e^{\int p(x) d x}$ を計算する. ここで $p(x)=-\frac{1}{x}$ より,
$$
e^{\int p(x) d x}=e^{\int \frac{2}{x} d x}=x^2
$$

となる. 両辺に積分因子をかけると，
$$
\begin{aligned}
\left(\frac{d z}{d x}+\frac{2 z}{x}\right) x^2 & =-2 x \\
\frac{d z}{d x} x^2-2 z x & =-2 x
\end{aligned}
$$

左辺に着目すると関数 $z x^2$ の微分になつているので両辺を積分し,左辺の積分定数 $C_1$ と右辺の積分定数 $C_2$ が存在するがまとめて $C=C_1+C_2$ とし, $z$ について整理したのち, $y$ について整理する.
$$
\begin{aligned}
\frac{d}{d x} z x^2 & =-2 x \\
z x^2 & =- x^2+C \\
y & =\frac{x^2}{-x^2+C}
\end{aligned}
$$

となる.

## 課題2
**課題2-1:** 次の微分方程式の一般解を求めよ．（手計算）

$$
y=\frac{dy}{dx}x + \left( \frac{dy}{dx}\right)^2 + 1
$$

この微分方程式は，クレロー型の微分方程式であるので $p=\frac{d y}{d x}$ として式を変形すると，
$$
y=p x+p^2+1
$$

このとき， $y$ と $p$ が $x$ の関数であることに注意して両辺を $x$ で微分する.
$$
\begin{aligned}
\frac{d y}{d x} & =p+x \frac{d p}{d x}+2 p \frac{d p}{d x} \\
p & =p+\frac{d p}{d x}(x+2 p) \\
0 & =\frac{d p}{d x}(x+2 p)
\end{aligned}
$$

この等式が成立するためには,
$$
\frac{d p}{d x}=\frac{d^2 y}{d x^2}=0
$$

もしくは
$$
x+2 p=0
$$

である必要がある. 前者の条件は2階微分すると0ということは導関数は定数，つまり
$$
\frac{d y}{d x}=p=C
$$

であるため, 一般解は
$$
y=C x+C^2+1
$$

となる. 後者の条件は, $p=-x / 2$ より, 式に代入すると，

$$
\begin{aligned}
y & =\left(-\frac{x}{2}\right) x+\left(-\frac{x}{2}\right)^2+1 \\
& =-\frac{1}{4} x^2+1
\end{aligned}
$$

という特異解が得られる.

**課題2-2:** 上記の微分方程式をsympyで実装し，一般解のグラフを描け．

```python
import numpy as np
from sympy import plotting
import matplotlib.pyplot as plt
from sympy import symbols, Eq, Derivative, Function, dsolve

x = symbols('x')
y = Function('y')(x)
dy = Derivative(y, x)
eq = Eq(y, dy*x + dy**2 + 1)
y_ = dsolve(eq, y)
display(y_)

N = 10

eqs = []
for c in np.linspace(-1, 1, N):
    eqs.append(y_.rhs.subs(symbols('C1'), c))
p = plotting.plot(*eqs, (x, -5, 5), show=False)

cm = plt.get_cmap('magma', N)
for i in range(cm.N):
    p[i].line_color = cm(i)
p.show()
```

## 課題3
**課題3-1:** 次の微分方程式が完全微分方程式であることを確かめよ．（手計算）

$$
3x \left ( 2y - x \right )dx + \left ( 3x^2 + 2y \right )dy = 0
$$

$p(x, y)=3 x(2 y-x), q(x, y)=3 x^2+2 y$ として $y$ と $x$ に関する偏微分をそれぞれ求めると，
$$
\begin{aligned}
\frac{\partial p(x, y)}{\partial y} & =\frac{\partial}{\partial y}\left(6 x y-3 x^2\right) \\
& =6 x \\
\frac{\partial q(x, y)}{\partial x} & =\frac{\partial}{\partial y}\left(3 x^2+2 y\right) \\
& =6 x
\end{aligned}
$$

となり， $\frac{\partial p(x, y)}{\partial y}=\frac{\partial q(x, y)}{\partial x}$ であるため完全微分方程式である.

**課題3-2:** 上記の微分方程式を解け．（手計算）

$p(x, y)$ を $x$ について積分すると，
$$
\begin{aligned}
f(x, y) & =\int p(x, y) d x+\phi(y) \\
& =\int(3 x(2 y-x)) d x+\phi(y) \\
& =\int\left(6 x y-3 x^2\right) d x+\phi(y) \\
& =-x^3+3 x^2 y+\phi(y)
\end{aligned}
$$

となる. 続いて、上記の $f(x, y)$ を $y$ について偏微分すると、
$$
\begin{aligned}
\frac{\partial f(x, y)}{\partial y} & =\frac{\partial}{\partial y}\left(-x^3+3 x^2 y+\phi(y)\right) \\
& =3 x^2+\frac{d}{d y} \phi(y)=q(x, y)
\end{aligned}
$$

したがつて， $\frac{d}{d y} \phi(y)=2 y$ とわかる. これを $y$ について積分すると、
$$
\begin{aligned}
\int \frac{d}{d y} \phi(y) & =\int 2 y d y \\
\phi(y) & =y^2
\end{aligned}
$$

したがって，任意定数 $C$ を用いて $f(x, y)=C$ を考慮し，以下の一般解が得られる.
$$
f(x, y)=-x^3+3 x^2 y+y^2=C
$$
