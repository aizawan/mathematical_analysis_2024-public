## 課題1
**課題1-1:** 次の微分方程式の一般解を求めよ．（手計算）

$$
\frac{d^2y}{dx^2} + 2\frac{dy}{dx} + \frac{1}{4}y = 0
$$


**解答**  
与えられた二階微分方程式の一般解を $y=e^{\lambda x}$ とおく. 特性方程式を考えると，
$$
\begin{aligned}
\frac{d^2}{d x^2} e^{\lambda x}+2 \frac{d}{d x} e^{\lambda x}+\frac{1}{4} e^{\lambda x} & =0 \\
\lambda^2 e^{\lambda x}+2 \lambda e^{\lambda x}+\frac{1}{4} e^{\lambda x} & =0 \\
\left(\lambda^2+2 \lambda+\frac{1}{4}\right) e^{\lambda x} & =0
\end{aligned}
$$

以上の方程式が得られ， $e^{\lambda x} \neq 0$ よりこの等式が満たされる条件は以下となる.
$$
\lambda^2+2 \lambda+\frac{1}{4}=0
$$

この方程式を $\lambda$ につて解くと，
$$
\lambda=-1 \pm \frac{\sqrt{3}}{2}
$$

であり，特性方程式は二つの異なる解を持つので，一般解は
$$
y=C_1 e^{-1+\frac{\sqrt{3}}{2}}+C_2 e^{-1-\frac{\sqrt{3}}{2}}
$$

となる．

**課題1-2:** 次の微分方程式の一般解を求めよ．（手計算）

$$
\frac{d^2y}{dx^2} + 4\frac{dy}{dx} + 4y = 0
$$


与えられた二階微分方程式の一般解を $y=e^{\lambda x}$ とおく. 特性方程式を考えると，

$$
\begin{aligned}
\frac{d^2}{d x^2} e^{\lambda x}+4 \frac{d}{d x} e^{\lambda x}+4 e^{\lambda x} & =0 \\
\lambda^2 e^{\lambda x}+4 \lambda e^{\lambda x}+4 e^{\lambda x} & =0 \\
\left(\lambda^2+4 \lambda+4\right) e^{\lambda x} & =0
\end{aligned}
$$

以上の方程式が得られ， $e^{\lambda x} \neq 0$ よりこの等式が満たされる条件は以下となる.
$$
\lambda^2+4 \lambda+4=0
$$

この方程式を $\lambda$ につて解くと，
$$
\lambda=-2
$$

の重根であった. したがって一般解は以下となる.
$$
y=\left(C_1 x+C_2\right) e^{-2 x}
$$

**課題1-3:** 次の微分方程式の一般解を求めよ．（手計算）

$$
\frac{d^2y}{dx^2} + \frac{dy}{dx} + y = 0
$$


解答
与えられた二階微分方程式の一般解を $y=e^{\lambda x}$ とおく. 特性方程式を考えると，
$$
\begin{aligned}
\frac{d^2}{d x^2} e^{\lambda x}+\frac{d}{d x} e^{\lambda x}+e^{\lambda x} & =0 \\
\lambda^2 e^{\lambda x}+\lambda e^{\lambda x}+e^{\lambda x} & =0 \\
\left(\lambda^2+\lambda+1\right) e^{\lambda x} & =0
\end{aligned}
$$

以上の方程式が得られ, $e^{\lambda x} \neq 0$ よりこの等式が満たされる条件は以下となる.
$$
\lambda^2+\lambda+1=0
$$

より、
$$
\lambda=-\frac{1}{2} \pm \frac{\sqrt{3}}{2} i
$$

であり共役な複素数解を持つ. したがつて一般解は以下となる.
$$
y=e^{-\frac{1}{2} x}\left(C_1 \cos \frac{\sqrt{3}}{2} x+C_2 \sin \frac{\sqrt{3}}{2} x\right)
$$

## 課題2
**課題2-1:** 次の微分方程式の一般解を求めよ．（手計算）

$$
\frac{d^2y}{dx^2} + 2\frac{dy}{dx} + y = 1
$$

解答非同次形の一般解は重ね合わせの性質から同次形の一般解と特殊解の和であった. そこで 与えられた微分方程式の同次形,
$$
\frac{d^2 y}{d x^2}+2 \frac{d y}{d x}+y=0
$$

の一般解を導出する. そのためには、一般解を $y=e^{\lambda x}$ としたときの特性方程式を考えればよ かった. 特性方程式は、
$$
\lambda^2+2 \lambda+1=(\lambda+1)^2=0
$$

より,
$$
\lambda=-1
$$

の重根となる. したがって，同次形の一般解が任意定数 $C_1, C_2$ を用いて，
$$
y_\alpha=\left(C_1 x+C_2\right) e^{-x}=C_1 x e^{-x}+C_2 e^{-x}
$$

となることを用いて,基本解は、
$$
y_1=x e^{-x}, \quad y_2=e^{-x}
$$

となる. 続いて，非同次形の特殊解を導出する. ここで非同次形での特殊解 $y_\beta$ が $x$ の関数項 $C_1(x), C_2(x)$ を持つ形式であると仮定する定数変化法を用いる.
$$
y_\beta=C_1(x) y_1(x)+C_2(x) y_2(x)
$$

定数変化法で仮定した関数項 $C_1(x), C_2(x)$ は以下の方程式を満たすことがわかっている.
$$
\left\{\begin{array}{l}
C_1^{\prime}(x) y_1(x)+C_2^{\prime}(x) y_2(x)=0 \\
C_1^{\prime}(x) y_1^{\prime}(x)+C_2^{\prime}(x) y_2^{\prime}(x)=g(x)
\end{array}\right.
$$

このとき $y_1=x e^{-x}, y_2=e^{-x}, y_1^{\prime}=e^{-x}-x e^{-x}, y_2^{\prime}=-e^{-x}, g(x)=1$ であることに注意して，上記の連立方程式を行列の形で書くと，
$$
\left(\begin{array}{cc}
x e^{-x} & e^{-x} \\
e^{-x}-x e^{-x} & -e^{-x}
\end{array}\right)\binom{C_1^{\prime}(x)}{C_2^{\prime}(x)}=\binom{0}{1}
$$

となる. これを $C_1^{\prime}(x), C_2^{\prime}(x)$ にいて解くと、
$$
\begin{aligned}
\binom{C_1^{\prime}(x)}{C_2^{\prime}(x)} & =\frac{1}{x e^{-x} \cdot-e^{-x}-e^{-x} \cdot\left(e^{-x}-x e^{-x}\right)}\left(\begin{array}{cc}
-e^{-x} & -e^{-x} \\
x e^{-x}-e^{-x} & x e^{-x}
\end{array}\right)\binom{0}{1} \\
& =-e^{2 x}\left(\begin{array}{cc}
-e^{-x} & -e^{-x} \\
x e^{-x}-e^{-x} & x e^{-x}
\end{array}\right)\binom{0}{1} \\
& =\binom{e^x}{-x e^x}
\end{aligned}
$$

となり，関数項 $C_1(x), C_2(x)$ の導関数は以下となる.
$$
\begin{aligned}
& C_1^{\prime}(x)=e^x \\
& C_2^{\prime}(x)=-x e^x
\end{aligned}
$$

これらを $x$ について積分する.

$$
\begin{aligned}
C_1(x) & =\int e^x d x=e^x+C_3 \\
C_2(x) & =\int-x e^x d x=e^x(1-x)+C_4
\end{aligned}
$$

ただし， $C_3, C_4$ は積分定数とする. 以上より，非同次形での特殊解 $y_\beta$ は
$$
\begin{aligned}
y_\beta & =\left(e^x+C_3\right) x e^{-x}+\left(e^x(1-x)+C_4\right) e^{-x} \\
& =x+C_3 x e^{-x}+(1-x) e^{-x}+C_4 e^{-x}
\end{aligned}
$$

となることがわかった. 最終的に求めたい非同次形の一般解 $y$ は同次形の一般解 $y_\alpha$ と非同次形の特殊解 $y_\beta$ の和であるので，
$$
\begin{aligned}
y & =y_\alpha+y_\beta \\
& =C_1 x e^{-x}+C_2 e^{-x}+x+C_3 x e^{-x}+(1-x) e^{-x}+C_4 e^{-x} \\
& =x+C_1 x e^{-x}+C_2 e^{-x}
\end{aligned}
$$

一般解は以上となる. ただし， $C_1, C_2$ はこれまで現れた任意定数 $C_1, C_2, C_3, C_4$ を再定義し た。

**課題2-2:** 次の微分方程式の一般解を求めよ．（手計算）

$$
\frac{d^2y}{dx^2} + 2\frac{dy}{dx} + y = e^{-x}
$$

解答非同次形の一般解は重ね合わせの性質から同次形の一般解と特殊解の和であった. そこで 与えられた微分方程式の同次形，
$$
\frac{d^2 y}{d x^2}+2 \frac{d y}{d x}+y=0
$$

の一般解を導出する. そのためには、一般解を $y=e^{\lambda x}$ としたときの特性方程式を考えればよ かった. 特性方程式は,
$$
\lambda^2+2 \lambda+1=(\lambda+1)^2=0
$$

より，
$$
\lambda=-1
$$

の重根となる. したがって、同次形の一般解が任意定数 $C_1, C_2$ を用いて,
$$
y_\alpha=\left(C_1 x+C_2\right) e^{-x}=C_1 x e^{-x}+C_2 e^{-x}
$$

となることを用いて,基本解は,
$$
y_1=x e^{-x}, \quad y_2=e^{-x}
$$

となる. 続いて，非同次形の特殊解を導出する. ここで非同次形での特殊解 $y_\beta$ が $x$ の関数項 $C_1(x), C_2(x)$ を持つ形式であると仮定する定数変化法を用いる.
$$
y_\beta=C_1(x) y_1(x)+C_2(x) y_2(x)
$$

定数変化法で仮定した関数項 $C_1(x), C_2(x)$ は以下の方程式を満たすことがわかっている.

$$
\left\{\begin{array}{l}
C_1^{\prime}(x) y_1(x)+C_2^{\prime}(x) y_2(x)=0 \\
C_1^{\prime}(x) y_1^{\prime}(x)+C_2^{\prime}(x) y_2^{\prime}(x)=g(x)
\end{array}\right.
$$

このとき $y_1=x e^{-x}, y_2=e^{-x}, y_1^{\prime}=e^{-x}-x e^{-x}, y_2^{\prime}=-e^{-x}, g(x)=e^{-x}$ であることに 注意して，上記の連立方程式を行列の形で書くと，
$$
\left(\begin{array}{cc}
x e^{-x} & e^{-x} \\
e^{-x}-x e^{-x} & -e^{-x}
\end{array}\right)\binom{C_1^{\prime}(x)}{C_2^{\prime}(x)}=\binom{0}{e^{-x}}
$$

となる. これを $C_1^{\prime}(x), C_2^{\prime}(x)$ について解くと，
$$
\begin{aligned}
\binom{C_1^{\prime}(x)}{C_2^{\prime}(x)} & =\frac{1}{x e^{-x} \cdot-e^{-x}-e^{-x} \cdot\left(e^{-x}-x e^{-x}\right)}\left(\begin{array}{cc}
-e^{-x} & -e^{-x} \\
x e^{-x}-e^{-x} & x e^{-x}
\end{array}\right)\binom{0}{e^{-x}} \\
& =-e^{2 x}\left(\begin{array}{cc}
-e^{-x} & -e^{-x} \\
x e^{-x}-e^{-x} & x e^{-x}
\end{array}\right)\binom{0}{e^{-x}} \\
& =\binom{1}{-x}
\end{aligned}
$$

となり、関数項 $C_1(x), C_2(x)$ の導関数は以下となる.
$$
\begin{aligned}
& C_1^{\prime}(x)=1 \\
& C_2^{\prime}(x)=-x
\end{aligned}
$$

これらを $x$ について積分する.
$$
\begin{aligned}
& C_1(x)=\int 1 d x=x+C_3 \\
& C_2(x)=\int-x d x=-\frac{x^2}{2}+C_4
\end{aligned}
$$

ただし， $C_3, C_4$ は積分定数とする．以上より，非同次形での特殊解 $y_\beta$ は
$$
\begin{aligned}
y_\beta & =\left(x+C_3\right) x e^{-x}+\left(-\frac{x^2}{2}+C_4\right) e^{-x} \\
& =x^2 e^{-x}+C_3 x e^{-x}-\frac{x^2}{2} e^{-x}+C_4 e^{-x}
\end{aligned}
$$

となることがわかった. 最終的に求めたい非同次形の一般解 $y$ は同次形の一般解 $y_\alpha$ と非同次形の特殊解 $y_\beta$ の和であるので,
$$
\begin{aligned}
y & =y_\alpha+y_\beta \\
& =C_1 x e^{-x}+C_2 e^{-x}+x^2 e^{-x}+C_3 x e^{-x}-\frac{x^2}{2} e^{-x}+C_4 e^{-x} \\
& =\frac{x^2}{2} e^{-x}+C_1 x e^{-x}+C_2 e^{-x}
\end{aligned}
$$

一般解は以上となる. ただし, $C_1, C_2$ はこれまで現れた任意定数 $C_1, C_2, C_3, C_4$ を再定義し た.
