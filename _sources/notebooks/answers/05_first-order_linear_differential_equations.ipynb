## 課題1
**課題1-1:** 次の微分方程式の一般解を求めよ．（手計算）

$$
\begin{align}
\frac{dy}{dx}+\cos(x)y=0
\end{align}
$$

解答

$$
\frac{d y}{y}=-\cos x d x
$$

$$
\begin{align}
\int \frac{d y}{y} & =\int-\cos x d x \\
\log |y| & =-\sin x+C^{\prime} \\
|y| & =e^{C^{\prime}} e^{-\sin x} \\
y & = \pm e^{C^{\prime}} e^{-\sin x} \\
y= \pm e^{C^{\prime}} \text { として, } & =C e^{-\sin x}
\end{align}
$$

## 課題2
**課題2-1:** 次の微分方程式の一般解を求めよ．（手計算）

$$
\begin{align}
\frac{dy}{dx}+(\sin x)y=\sin x
\end{align}
$$

解答

積分因子法で解く．

$$
e^{\int p(x) d x}=e^{\int \sin x d x} = e^{-\cos x} \\
$$

$$
e^{-\cos x} \frac{d y}{d x}+e^{-\cos x}(\sin x) y=e^{-\cos x} \sin x
$$

$$
\frac{d}{d x}\left(y e^{-\cos x}\right)=e^{-\cos x} \sin x
$$

$$
y e^{-\cos x}=e^{-\cos x}+C
$$

$$
y=1+C e^{\cos x}
$$

$$
\begin{align}
\frac{dy}{dx}+\frac{y}{x}=x
\end{align}
$$

$$

$$

$$
\begin{align}
\left(\frac{d y}{d x}+\frac{y}{x}\right) x & =x^2 \\
\frac{x d y}{d x}+y & =x^2
\end{align}
$$

$$
\begin{align}
\frac{d}{d x} x y & =x^2 \\
x y & =\frac{1}{3} x^3+C \\
y & =\frac{1}{3} x^2+\frac{C}{x}
\end{align}
$$

## 課題4
**課題4-1:** 次の微分方程式の一般解を定数変化法で求めよ．（手計算）

$$
\begin{align}
\frac{dy}{dx}+\frac{y}{x}=x
\end{align}
$$

まず， $\frac{d y}{d x}+\frac{y}{x}=0$ の一般解を考えると，
$$
y_1=\frac{C_1}{x}
$$

となることから,特殊解を次のように仮定する.
$$
y_2=\frac{C(x)}{x}
$$

$C^{\prime}(x)=q(x) e^{\int p(x) d x}$ より, 

$$
\begin{align}
C^{\prime}(x)&=x e^{\int \frac{1}{x} d x} \\
C^{\prime}(x)&=x^2 d x
\end{align}
$$

これを $x$ で積分すると,

$$
\begin{align}
\int C^{\prime}(x) & =\int x^2 d x \\
C(x) & =\frac{1}{3} x^3+C_2
\end{align}
$$

となる. これを仮定した特殊解 $y_2$ に代入すると,
$$
y_2=\frac{\frac{1}{3} x^3+C_2}{x}
$$

となる. 最後に、非同次形の一般解 $y$ は同次形の一般解 $y_1+$ 非同次形の特殊解 $y_2$ だったので
$$
\begin{align}
& y=y_1+y_2 \\
& y=\frac{C_1}{x}+\frac{\frac{1}{3} x^3+C_2}{x} \\
& y=\frac{1}{3} x^2+\frac{C}{x}
\end{align}
$$

一般解は上式となる. ただし任意定数 $C=C_1+C_2$ している.

**課題4-1:** 上記の微分方程式をsympyで実装し，一般解のグラフを描け．

```python
# 解答用セル

import numpy as np
from sympy import plotting
import matplotlib.pyplot as plt
from sympy import symbols, Eq, Derivative, Function, dsolve

x = symbols('x')
y = Function('y')(x)
dy = Derivative(y, x)

eq = Eq(dy + y/x, x)
y_ = dsolve(eq, y)
display(y_)

N = 10

eqs = []
for c in np.linspace(-3, 3, N):
    eqs.append(y_.rhs.subs(symbols('C1'), c))
p = plotting.plot(*eqs, (x, -5, 5), ylim=(-5, 5), show=False)

cm = plt.get_cmap('magma', N)
for i in range(cm.N):
    p[i].line_color = cm(i)
p.show()
```
